### Mini batch

- batch size : `100` per one-batch

#### \# DataLoader
- 원래의 DataSet을 원하는 `batch_size`로 분할하는 것
    - `train_loader`는 60000개의 data를 한 batch 당 100장씩 갖도록 나누므로 600개의 batch-set을 갖는다.
    - `test_loader`는 10000개의 data를 한 batch 당 100장씩 갖도록 나누므로 100개의 batch-set을 갖는다.

    
- DataLoader로 나눈다고 해서, 바로 데이터가 들어가있는 sets를 형성하는 것이 아니다.
    - `shuffle=True` : 각각의 batch-set에 매번 다른 data들이 다시 섞여서 들어간다. 즉, 600개의 batch-set의 data들은 계속 바뀐다.
    - `shuffle=False` : batch-set을 한 번 할당하면 다시 바뀌지 않는다. 

```python
batch_size = 100
train_loader = torch.utils.data.DataLoader(mnist_train, batch_size=batch_size, shuffle=True, num_workers=2, drop_last=True)
test_loader = torch.utils.data.DataLoader(mnist_test, batch_size=batch_size, shuffle=False, num_workers=2, drop_last=True)
```

```python
len(train_loader), len(test_loader)
```

```python
hidden_size = 10

class Encoder(nn.Module):
    def __init__(self):
        super(Encoder, self).__init__()
        self.fc1 = nn.Sequential(
                        nn.Conv2d(2, 16, 3, padding=1),   # (입력채널수, 출력채널수, 필터크기) batch x 16 x 28 x 28
                        nn.BatchNorm2d(16),
                        nn.ReLU(),
                        nn.MaxPool2d(2, 2),               # (필터크기, stride)
            
                        nn.Conv2d(16, 32, 3, padding=1),  # (입력채널수, 출력채널수, 필터크기) batch x 32 x 14 x 14
                        nn.BatchNorm2d(32),
                        nn.ReLU(),
                        nn.MaxPool2d(2, 2),               # (필터크기, stride)
            
                        nn.Conv2d(32, 32, 3, padding=1),  # (입력채널수, 출력채널수, 필터크기) batch x 32 x 7 x 7
                        nn.BatchNorm2d(32),
                        nn.ReLU(),
                        nn.MaxPool2d(2, 2),
            
                        nn.Conv2d(32, 64, 3),             # (입력채널수, 출력채널수, 필터크기) batch x 64 x 1 x 1
                        nn.BatchNorm2d(64),
                        nn.ReLU(),
        )
        self.fc2_1 = nn.Sequential(
                        nn.Linear(64*1*1, 256),
                        nn.ReLU(),
                        nn.Linear(256, hidden_size),
        )
        self.fc2_2 = nn.Sequential(
                        nn.Linear(64*1*1, 256),
                        nn.ReLU(),
                        nn.Linear(256, hidden_size),
        )
        self.relu = nn.ReLU()
                 
            
    def encode(self, x, label):
        
        # label을 x의 shape에 맞게 expand 시키기
        # x.shape = (100, 1, 28, 28)
        # lable.shape = (100) -> (100, 1, 28, 28)
        c = label.view(-1, 1, 1, 1).expand_as(x)
        
        # x와 label을 concatenate 시키기
        # input_.shape = (100, 2, 28, 28)
        input_ = torch.cat((x, c.float()), 1)
        
        # input_을 fc1에 입력
        # out.shape = (100, 64, 1, 1)
        out = self.fc1(input_) 
        
        # out.shape = (100, 64)
        out = out.view(x.size(0), -1)
        out = self.relu(out)
        
        # mu와 log_var 생성
        mu = self.fc2_1(out)
        log_var = self.fc2_2(out)
        
        return mu, log_var
    
    
    def forward(self, x, label):

        # mu, log_var 출력
        mu, log_var = self.encode(x, label)
        # noise 생성
        noise = torch.randn(mu.size()).cuda()
        # z 생성
        z = mu + log_var.exp() * noise    
        
        return z, mu, log_var
        
encoder = Encoder().cuda()
```

### 코드 뜯어보기

```python
fc1 = nn.Sequential(
            nn.Conv2d(2, 16, 3, padding=1),
            nn.BatchNorm2d(16),
            nn.ReLU(),
            nn.MaxPool2d(2,2), 
            nn.Conv2d(16,32,3,padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2,2),
            nn.Conv2d(32,32,3,padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2,2),
            nn.Conv2d(32,64,3), 
            nn.BatchNorm2d(64),
            nn.ReLU(),
)

fc2_1 = nn.Sequential(
            nn.Linear(64*1*1, 256),
            nn.ReLU(),
            nn.Linear(256, hidden_size),
)

fc2_2 = nn.Sequential(
            nn.Linear(64*1*1, 256),
            nn.ReLU(),
            nn.Linear(256, hidden_size),
)

relu = nn.ReLU()
```

```python
for idx, [image, label] in enumerate(train_loader):
    x = image
    print("[x]\t : ", x.shape)
    label = label
    print("[label]\t : ", label.shape)
    c = label.view(-1, 1, 1, 1).expand_as(x)
    print("[c]\t : ", c.shape)
    input_ = torch.cat((x, c.float()), 1)
    print("[input]\t : ", input_.shape)
    out = fc1(input_)
    out = out.view(x.size(0), -1)
    print("[out]\t : ", out.shape)
    
    mu = fc2_1(out)
    print("[mu]\t : ", mu.shape)
    log_var = fc2_2(out)
    print("[logvar] : ", log_var.shape)
    
    noise = torch.randn(mu.size())
    z = mu + log_var.exp() * noise    
    print("[z]\t : ", z.shape)
    break
```

```python
class Decoder(nn.Module):
    def __init__(self):
        super(Decoder,self).__init__()
        self.fc1 = nn.Sequential(
                        nn.Linear(hidden_size+1, 800),
                        nn.BatchNorm1d(800),
                        nn.ReLU(),
                        nn.Linear(800,1568),
                        nn.ReLU(),
        )
        self.fc2 = nn.Sequential(
                        nn.ConvTranspose2d(32+1,16,3,2,1,1),
                        nn.ReLU(),
                        nn.BatchNorm2d(16),
                        nn.ConvTranspose2d(16,8,3,2,1,1),
                        nn.ReLU(),
                        nn.BatchNorm2d(8),
                        nn.ConvTranspose2d(8,1,3,1,1),
                        nn.BatchNorm2d(1),
        )
        self.sigmoid = nn.Sigmoid()
        self.relu = nn.ReLU()
    
    def forward(self, x, label):
       
        out = self.fc1(torch.cat((x,label.unsqueeze(1)),1))
        out = self.relu(out)
        out = out.view(x.size(0),32,7,7)
        
        c = label.view(-1,1,1,1).expand(x.size(0),1,7,7)
        out = self.fc2(torch.cat((out,c),1))
        out = self.sigmoid(out)
        out = out.view(x.size(0),1,28,28)
        
        return out
        
decoder = Decoder().cuda()
```

```python
reconstruction_function = nn.BCELoss(size_average=False)

def loss_function(recon_x, x, mu, logvar):
    BCE = reconstruction_function(recon_x, x)

    # see Appendix B from VAE paper:
    # Kingma and Welling. Auto-Encoding Variational Bayes. ICLR, 2014
    # https://arxiv.org/abs/1312.6114
    # 0.5 * sum(1 + log(sigma^2) - mu^2 - sigma^2)
    KLD_element = mu.pow(2).add_(logvar.exp()).mul_(-1).add_(1).add_(logvar)
    KLD = torch.sum(KLD_element).mul_(-0.5)

    return BCE + KLD

parameters = list(encoder.parameters()) + list(decoder.parameters())
optimizer = torch.optim.Adam(parameters, lr=0.00001)
```

```python
%%time
try:
    encoder, decoder = torch.load('./conditional_vae.pkl')
    print("\n--------model restored--------\n")
except:
    print("\n--------model not restored--------\n")
    pass

num_epoch = 500

for i in range(num_epoch):
    for j, [image, label] in enumerate(train_loader):
        image = image.cuda()
        label = label.cuda() 
        
        z, mu, log_var  = encoder(image, label.float())
        output = decoder(z, label.float())
        
        loss = loss_function(output, image, mu, log_var)
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    print("{}: loss is".format(i+1), loss.cpu().item())       
```

### 한 latent value z에 각기 다른 label을 input으로 주었을 때
